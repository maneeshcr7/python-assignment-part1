```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
```

```python
students = []

for student in raw_students:
    # Clean the data
    cleaned_name = student["name"].strip().title()
    cleaned_roll = int(student["roll"])
    cleaned_marks = [int(mark) for mark in student["marks_str"].split(", ")]
    
    # Validate name
    name_words = cleaned_name.split()
    is_valid = all(word.isalpha() for word in name_words)
    validity = "✓ Valid name" if is_valid else "✗ Invalid name"
    
    print(f"{cleaned_name}: {validity}")
    
    # Print profile card
    print("================================")
    print(f"Student : {cleaned_name}")
    print(f"Roll No : {cleaned_roll}")
    print(f"Marks   : {cleaned_marks}")
    print("================================")
    print()
    
    students.append({"name": cleaned_name, "roll": cleaned_roll, "marks": cleaned_marks})

# Find and print student with roll number 103
student_103 = next(s for s in students if s["roll"] == 103)
print(f"Student 103 - ALL CAPS: {student_103['name'].upper()}, lowercase: {student_103['name'].lower()}")
```

```python
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]
```

```python
def get_grade(score):
    if 90 <= score <= 100:
        return "A+"
    if 80 <= score <= 89:
        return "A"
    if 70 <= score <= 79:
        return "B"
    if 60 <= score <= 69:
        return "C"
    return "F"

print("Subject\tMarks\tGrade")
for subj, m in zip(subjects, marks):
    grade = get_grade(m)
    print(f"{subj}\t{m}\t{grade}")

total_marks = sum(marks)
average_marks = round(total_marks / len(marks), 2)
max_idx = max(range(len(marks)), key=lambda i: marks[i])
min_idx = min(range(len(marks)), key=lambda i: marks[i])
highest_subject = subjects[max_idx]
highest_marks = marks[max_idx]
lowest_subject = subjects[min_idx]
lowest_marks = marks[min_idx]

print(f"\nTotal marks: {total_marks}")
print(f"Average marks: {average_marks}")
print(f"Highest scoring subject: {highest_subject} ({highest_marks})")
print(f"Lowest scoring subject: {lowest_subject} ({lowest_marks})")

new_count = 0
while True:
    new_sub = input("Enter subject name (or 'done' to finish): ").strip()
    if new_sub.lower() == "done":
        break
    mark_str = input(f"Enter marks for {new_sub} (0–100): ").strip()
    try:
        new_mark = float(mark_str)
        if not (0 <= new_mark <= 100):
            print("Invalid marks: must be between 0 and 100. Entry ignored.")
            continue
    except ValueError:
        print("Invalid input: marks must be numeric. Entry ignored.")
        continue

    subjects.append(new_sub)
    marks.append(new_mark)
    new_count += 1
    print(f"Added: {new_sub} ({new_mark})")

updated_total = sum(marks)
updated_avg = round(updated_total / len(marks), 2) if marks else 0
print(f"\nNew subjects added: {new_count}")
print(f"Updated total subjects: {len(marks)}")
print(f"Updated average across all subjects: {updated_avg}")
```

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]
```

```python
# use existing class_data
print("Name              | Average | Status")
print("----------------------------------------")

pass_count = 0
fail_count = 0
averages = []

for name, marks in class_data:
    avg = round(sum(marks) / len(marks), 2)
    status = "Pass" if avg >= 60 else "Fail"
    if status == "Pass":
        pass_count += 1
    else:
        fail_count += 1
    averages.append(avg)
    print(f"{name:<17} | {avg:7.2f} | {status}")

class_avg = round(sum(averages) / len(averages), 2)
topper_idx = max(range(len(averages)), key=lambda i: averages[i])
topper_name = class_data[topper_idx][0]
topper_avg = averages[topper_idx]

print()
print(f"Passed: {pass_count}")
print(f"Failed: {fail_count}")
print(f"Class topper: {topper_name} ({topper_avg:.2f})")
print(f"Class average: {class_avg:.2f}")
```

```python
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "
```

```python
# Step 1: Strip leading and trailing whitespace from essay
clean_essay = essay.strip()

# Step 2: Convert clean_essay to Title Case and print it
print("Title Case:")
print(clean_essay.title())

# Step 3: Count how many times the word "python" appears in clean_essay (case-insensitive)
count_python = clean_essay.lower().count("python")
print(f"\nCount of 'python' (case-insensitive): {count_python}")

# Step 4: Replace every occurrence of "python" in clean_essay with "Python 🐍" and print the result
replaced_essay = clean_essay.lower().replace("python", "Python 🐍")
print("\nReplaced essay:")
print(replaced_essay)

# Step 5: Split clean_essay into individual sentences by splitting on ". " (a period followed by a space)
sentences = clean_essay.split(". ")
print("\nSentences list:")
print(sentences)

# Step 6: Print each sentence on a new line, numbered starting from 1. Add a "." at the end if it does not already end with one.
print("\nNumbered sentences:")
for i, sentence in enumerate(sentences, start=1):
    if not sentence.endswith("."):
        sentence += "."
    print(f"{i}. {sentence}")
```
